```python
import pandas as pd

# Read the two datasets
df_stringency = pd.read_csv('data/map/stringency_average_1.csv')
df_containment = pd.read_csv('data/map/containement_average_1.csv')

# Merge the datasets on the 'location' column
merged_df = pd.merge(df_stringency, df_containment, on='location')
merged_df
```

```python
# Calculate the ratio between stringency index and containment index
merged_df['ratio'] = (merged_df['average_stringency_index'] + merged_df['average_containment_index'])/2

# Save the merged dataset with the ratio as a new CSV file
merged_df.to_csv('data/map/merged_data.csv', index=False)
```

```python
import csv
import pandas as pd

df1 = pd.read_csv('data/owid-covid-data.csv')
df2 = pd.read_csv('data/covid-containment-and-health-index.csv')
```

```python
# Filter the DataFrame based on your conditions
countries = ['Belgium', 'France', 'Germany', 'Italy', 'Luxembourg', 'Netherlands', 'Denmark', 'Ireland', 'Greece', 'Portugal', 'Spain', 'Austria', 'Finland', 'Sweden', 'Cyprus', 'Estonia', 'Latvia', 'Lithuania', 'Malta', 'Poland', 'Czechia', 'Slovakia', 'Slovenia', 'Hungary', 'Bulgaria', 'Romania', 'Croatia']

#'2020-02-15' <= row['date'] <= '2020-07-20' 1st wave
#'2020-07-20' <= row['date'] <= '2021-07-10' 2nd wave
#'2021-07-10' <= row['date'] <= '2022-12-31' 3rd wave

start_date = '2021-07-10'
end_date = '2023-02-16'

date_condition1 = (df1['date'] >= start_date) & (df1['date'] <= end_date)
date_condition2 = (df2['date'] >= start_date) & (df2['date'] <= end_date)
stringency_condition = df1['stringency_index'].notnull()
containement_condition = df2['containment_index'].notnull()

filtered_df1 = df1[(df1['location'].isin(countries)) & date_condition1 & stringency_condition]
filtered_df2 = df2[(df2['location'].isin(countries)) & date_condition2 & containement_condition]

print(filtered_df1)
print(filtered_df2)

filtered_df2.drop('date', axis=1, inplace=True)

# Merge the datasets on the 'location' column
merged_df = pd.merge(filtered_df1, filtered_df2, on='location')
merged_df
```
